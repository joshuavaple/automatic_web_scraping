```python
pages = [i for i in range(1, 11)]
urls = ['https://www.who.int/emergencies/disease-outbreak-news/' + str(page) for page in pages]
hrefs = []

for url in urls:
    # Send a GET request to the URL
    response = requests.get(url, verify=False)
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')
    # Find all anchor tags (i.e., <a> tags) in the HTML content
    for link in soup.find_all('a', {'class': 'sf-list-vertical__item'}):
        href = link.get('href')
        hrefs.append(href)
```

```python
def getArticleTexts(urls: list):
    results = {'title':[], 'body':[], 'url':[]}
    for url in urls:
        results['url'].append(url)

        response = requests.get(url, verify = False)
        # Use BeautifulSoup to parse the HTML content of the response
        soup = BeautifulSoup(response.content, 'html.parser')
        h1_tags = soup.find_all('h1')
        header_text = ''
        for h1_tag in h1_tags:
            header_text += h1_tag.get_text()
        results['title'].append(header_text)

        article_texts = ''
        # Find the article tag with class attribute "sf-detail-body-wrapper"
        article = soup.find('article', class_='sf-detail-body-wrapper')
        # Find all p tags under the article tag and print their text
        for p in article.find_all('p'):
            article_texts += (p.text)
        results['body'].append(article_texts)
        
    return results
```

```python
results = getArticleTexts(urls = hrefs)
```

```python
df_articles = pd.DataFrame(results)
df_articles.to_excel('who.xlsx')
```
